```python
#9.Display all duplicate items from the list:10,20,60,30,20,40,30,60,70,80
list2=[10,20,60,30,20,40,30,60,70,80]
uniqueList = []
duplicateList = []
for i in list2:
    if i not in uniqueList:
        uniqueList.append(i)
    elif i not in  duplicateList:
            duplicateList.append(i)
print("The  given list is list2 = [10,20,60,30,20,40,30,60,70,80]\n ")        
print( "The duplicate items in list2 is :",duplicateList)
```

```python

#7.Write a python program to construct the following pattern,using a nested for loop
n=5;
for i in range(n):
    for j in range(i):
        print('* ',end="")
    print('')
for i in range(n,0,-1):
    for j in range(i):
        print('* ',end="")
    print('')
```
